```python
import pandas as pd
from shared.teaching_model import build_all_sections, build_all_course_schedules, build_master_schedule

schedules = build_all_course_schedules(
    sections_file="data/f2026_sections.csv",
)

rows = []

for schedule in schedules:
    for item in schedule["weeks"]:
        rows.append({
            "Course": schedule["course_code"],
            "Section": schedule["section"],
            "DateValue": item["date"],
            "Date": item["date_str"],
            "Type": item["type"],
            "Week": item.get("week"),
            "Topic": item.get("topic"),
            "Readings": item.get("readings"),
            "Assignments": item.get("assignments"),
            "Time": item.get("class_time"),
            "Location": item.get("location"),
        })

df = pd.DataFrame(rows)

course_code = "MOS 3321"

course_df = (
    df[
        df["Course"].str.replace(" ", "")
        == course_code.replace(" ", "")
    ]
    .sort_values(["Section", "DateValue"])
    .drop(columns=["DateValue"])
    .reset_index(drop=True)
)

course_df
```
