```python
from shared.brightspace import generate_course_info, generate_schedule_page
from shared.teaching_model import build_all_course_terms

course_terms = build_all_course_terms()

course_term = next(
    term
    for term in course_terms
    if term["course_code"].replace(" ", "") == "MOS2320"
)

# Generate the shared course information page
course_info = generate_course_info(course_term)

with open("course_information.html", "w", encoding="utf-8") as f:
    f.write(course_info)

# Generate one schedule page per section
for schedule in course_term["schedules"]:
    html = generate_schedule_page(schedule)

    with open(f"schedule_{schedule['section']}.html", "w", encoding="utf-8") as f:
        f.write(html)

    print(schedule["section"])
```
