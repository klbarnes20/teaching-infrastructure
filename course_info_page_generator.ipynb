```python
from shared.brightspace import generate_course_info, generate_schedule_page
from shared.teaching_model import build_all_course_terms

course_terms = build_all_course_terms()

course_term = next(
    term
    for term in course_terms
    if term["course_code"].replace(" ", "") == "MOS2320"
)

# Generate the shared course information page
course_info = generate_course_info(course_term)

with open("course_information.html", "w", encoding="utf-8") as f:
    f.write(course_info)

# Generate one schedule page per section
for schedule in course_term["schedules"]:
    html = generate_schedule_page(schedule)

    with open(f"{schedule['course_code']}-schedule_{schedule['section']}.html", "w", encoding="utf-8") as f:
        f.write(html)
    print(schedule["section"])
```

```python
import pandas as pd
from shared.teaching_model import build_all_course_schedules

schedules = build_all_course_schedules(
    sections_file="data/f2026_sections.csv"
)

section_551 = next(
    schedule
    for schedule in schedules
    if (
        schedule["course_code"].replace(" ", "") == "MOS3321"
        and str(schedule["section"]) == "551"
    )
)

existing_dates = {
    item["date"]
    for item in section_551["weeks"]
}

for exception in section_551.get("schedule_exceptions", []):
    if exception.get("type") == "review":
        exception_date = pd.to_datetime(
            exception["date"]
        ).date()

        applicable_sections = [
            str(section)
            for section in exception.get("sections", [])
        ]

        print("Exception:", exception)
        print("Current section:", repr(str(section_551["section"])))
        print("Applicable sections:", applicable_sections)
        print("Date:", exception_date)
        print("Date already present:", exception_date in existing_dates)
        print(
            "Should be added:",
            str(section_551["section"]) in applicable_sections
            and exception_date not in existing_dates
            and not exception.get("consumes_week", True)
        )
```
