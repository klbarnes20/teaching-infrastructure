```python

from generators.brightspace import render_course_basics, render_learning_outcomes, render_course_policies, render_assessment_table, render_textbook
from shared.course_schedule import build_all_course_schedules
import os

output_file = os.path.abspath("test.html")

all_schedules = build_all_course_schedules(
    sections_file="data/f2026_sections.csv",
)

config = next(
    schedule
    for schedule in all_schedules
    if schedule["course_code"].replace(" ", "") == "MOS3321"
)

basics = render_course_basics(config)
outcomes = render_learning_outcomes(config)
policies = render_course_policies(config)
assessments = render_assessment_table(config)
textbook = render_textbook(config)
html = basics + outcomes + textbook + assessments + policies

print(html)


with open("test.html", "w", encoding="utf-8") as f: 
    f.write(html)

print("saved to:", output_file)
os.startfile(output_file)
```
