```python
######## GENERATOR 

rows = []

# ==========================================
# UNIVERSAL TASKS
# ==========================================

current_section = None
#term_name = section["term"].get("term_label", "Fall 2026")

for task in UNIVERSAL_TASKS:

    # Skip midterm tasks for courses without a midterm
    if task["section"] == "Midterm" and not section.get("has_midterm", True):        continue

    # Add section row when section changes
    if task["section"] != current_section:

        rows.append({
            "TYPE": "section",
            "CONTENT": task["section"]
        })

        current_section = task["section"]

    due_date = (
        get_anchor_date(section, task["anchor"])
        + timedelta(days=task["offset"])
    )

    rows.append({
        "TYPE": "task",
        "CONTENT": task["task"],
        "DESCRIPTION": task.get("description", ""),
        "PRIORITY": task.get("priority", 1),
        "INDENT": task.get("indent", 1),
        "LABELS": ",".join(task.get("labels", [])),
        "DATE": due_date.strftime("%Y-%m-%d")
    })


# ==========================================
# WEEKLY TASKS
# ==========================================

rows.append({
    "TYPE": "section",
    "CONTENT": "Weekly Tasks"
})

weeks = section["weeks"]

weekly_task_date_offsets = {
    "Post weekly announcement": 0,
    "Prepare lecture slides": -1,
    "Review readings": -1,
    "Prepare Poll Everywhere": -1,
    "Mark participation": 1,
}

for week in weeks:

    date_value = week["date"]

    if week.get("type") == "midterm":
        title = "Midterm Exam"
    else:
        title = f"Week {week.get('week')}: {week.get('topic', 'TBD')}"

    rows.append({
        "TYPE": "task",
        "CONTENT": title,
        "DESCRIPTION": "",
        "PRIORITY": 3,
        "INDENT": 1,
        "LABELS": week.get("type", "weekly"),
        "DATE": date_value.strftime("%Y-%m-%d")
    })

    for task in WEEKLY_TASKS:
        offset = weekly_task_date_offsets.get(task["task"], 0)

        task_date = date_value + timedelta(days=offset)

        rows.append({
            "TYPE": "task",
            "CONTENT": task["task"],
            "DESCRIPTION": task.get("description", ""),
            "PRIORITY": task.get("priority", 2),
            "INDENT": 2,
            "LABELS": ",".join(task.get("labels", [])),
            "DATE": task_date.strftime("%Y-%m-%d")
        })

    readings = week.get("readings", [])

    if isinstance(readings, str):
        readings = [readings]

    if readings:
        rows.append({
            "TYPE": "task",
            "CONTENT": f"Review readings: {', '.join(readings)}",
            "DESCRIPTION": "",
            "PRIORITY": 2,
            "INDENT": 2,
            "LABELS": "readings",
            "DATE": (date_value - timedelta(days=1)).strftime("%Y-%m-%d")
        })

    assignments = week.get("assignments", []) or []

    if isinstance(assignments, str):
        assignments = [assignments]

    for assignment in assignments:
        rows.append({
            "TYPE": "task",
            "CONTENT": f"Assignment Due: {assignment}",
            "DESCRIPTION": "",
            "PRIORITY": 3,
            "INDENT": 2,
            "LABELS": "assignment",
            "DATE": date_value.strftime("%Y-%m-%d")
        })

        rows.append({
            "TYPE": "task",
            "CONTENT": f"Grade: {assignment}",
            "DESCRIPTION": "",
            "PRIORITY": 3,
            "INDENT": 2,
            "LABELS": "grading",
            "DATE": (date_value + timedelta(days=7)).strftime("%Y-%m-%d")
        })
# ==========================================
# DATAFRAME
# ==========================================

import pandas as pd

df = pd.DataFrame(rows)

display(df.head(75))

course_code = section["course_code"].replace(" ", "")
section_num = section["section"]

df.to_csv(
    f"outputs/todoist/{course_code}_{section_num}_todoist.csv",
    index=False
)
```
